```python
import socket
import time

```

```python
s = socket.socket()
s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
s.bind(('',3334))
s.listen()

session , addr = s.accept()
print(session)
```

```python
i=0
while i<10:
    data = session.recv(1500)
    print(i,data)
    i+=1
    time.sleep(3)
    
```
